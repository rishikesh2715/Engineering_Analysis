The gradient of the function \$(f = xyz\)$ in cylindrical coordinates \((\rho, \phi, z)\) is:

\[
\nabla f = 2\rho z \cos(\phi) \sin(\phi) \hat{\rho} + \rho z (-\sin^2(\phi) + \cos^2(\phi)) \hat{\phi} + \rho^2 \cos(\phi) \sin(\phi) \hat{z}
\]

Where:

- \(\hat{\rho}\), \(\hat{\phi}\), and \(\hat{z}\) are the unit vectors in the direction of increasing \(\rho\), \(\phi\), and \(z\) respectively.
- \(f\) is expressed in cylindrical coordinates as \(f(\rho, \phi, z) = \rho^2 z \cos(\phi) \sin(\phi)\).


```python
from sympy import symbols, Eq, solve, Matrix

pi1, pi2, pi3 = symbols('pi1 pi2 pi3')

# transition matrix
P = Matrix([
    [1/2, 1/4, 1/4],
    [1/3, 0, 2/3],
    [1/2, 1/2, 0]
])
print(P)
eqs = [Eq(pi1, P[0,0]*pi1 + P[1,0]*pi2 + P[2,0]*pi3),
       Eq(pi2, P[0,1]*pi1 + P[1,1]*pi2 + P[2,1]*pi3),
       Eq(pi3, P[0,2]*pi1 + P[1,2]*pi2 + P[2,2]*pi3),
       Eq(pi1 + pi2 + pi3, 1)]

solution = solve(eqs, (pi1, pi2, pi3))
print(solution)
sum(solution.values())

```

```python
import numpy as np
import matplotlib.pyplot as plt

particles = 10_000
particles_locations = np.zeros(particles)
time_steps = [50, 500]

for time in range(1, max(time_steps) + 1 ):
    steps = np.random.choice([-0.1, 0.1], size = particles)
    particles_locations += steps

    if time in time_steps:
        plt.figure(figsize = (10,8))
        plt.hist(particles_locations, bins = 100, density = True, alpha = 0.5, label = f'{time} time steps')
        plt.xlim(-3,3)
        plt.xlabel('position')
        plt.ylabel('Number of particles')
        plt.title('Random walk of particles')
        plt.legend()
        plt.show()
```
